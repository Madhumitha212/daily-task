```python
import numpy as np
import pandas as pd 

res = ["PizzaHut", "Aha Biryani", "KFC king", "kottayam", "Murugan"] 
price = np.random.randint(50,2000,350)
delivery_time = np.random.randint(-1,60,350)
customer_rating = np.random.randint(1,5,350)


data = {  
    "order_id" : np.arange(1,351,1),
    "restaurant" : np.random.choice(res, 350),
    "price": price,
    "delivery_time" : delivery_time,
    "customer_rating" : customer_rating
}
df = pd.DataFrame(data)
df
```

```python
poor_performing = df["customer_rating"]<2
poor_performing
df[poor_performing]
```

```python
revenue = df.groupby(df["restaurant"])["price"].sum()
revenue
```

```python
fifty_percentage = np.percentile(delivery_time,50)
fifty_percentage
```

```python
delivery_below_fifty = df[df["delivery_time"]<31]
delivery_below_fifty
```

```python
df[df["delivery_time"]<3]
```

```python
df['delivery_time'] = df['delivery_time'].mask(df['delivery_time'] < 3, df['delivery_time'].median())
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["delivery_time"],df["customer_rating"])
plt.xlabel("Delivery Time")
plt.ylabel("Customer Rating")
plt.title("Delivery time Vs customer ratings")
plt.grid()
plt.show()
```
