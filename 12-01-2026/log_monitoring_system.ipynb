```python
import numpy as np

list1 = []
with open("sample_text.txt","r") as file:
    for line in file:
        list1.append(int(line))
array = np.array(list1).reshape(24,60)
array.shape
sum_requests = array.sum(axis = 1)
sum_requests
print(sum_requests)
peak_hours = np.argsort(sum_requests)
print("peak_hours:",peak_hours[-2:])
window = 3 
moving_avg = np.convolve(sum_requests, np.ones(window)/window, mode="valid")
print("Smoothed moving average",moving_avg)
```

```python
import pandas as pd 
import numpy as np 
import random
from faker import Faker

faker = Faker()
n = 1440
endpoints = ["login", "sigup"]
timestamps = timestamps = pd.date_range("2025-01-01", periods=1440, freq="min")
endpoint_choices = [random.choice(endpoints) for _ in range(n)] 
response_times = [random.randint(100, 1000) for _ in range(n)] 
df = pd.DataFrame({ "timestamp": timestamps, "endpoint": endpoint_choices, "response_time": response_times }) 
threshold = 500
avg_response = df.groupby("endpoint")["response_time"].mean()
print(avg_response)
end_points = avg_response[avg_response > threshold]
print(end_points)
df = df.set_index("timestamp") 
windowed = df.resample("5min")["response_time"].mean() 
degraded = windowed[windowed > threshold]
degraded
```
